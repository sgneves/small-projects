## Word embeddings

We will use word embeddings so that our model can associate our emojis with a wider range of words. Even if the training set explicitly relates only a few words to a particular emoji, the algorithm can generalize and associate different words in the test set to the same emoji. By using word embeddings, we can build an accurate classifier with even a small training set. We will use 50-dimensional [Glove word vectors](https://nlp.stanford.edu/projects/glove/) trained with a corpus of 6 billion words and a 400,000 vocabulary.

```python
# Read the pre-trained GloVe word embeddings
embeddings_vecs = {}

with open('glove.6B.50d.txt', 'r', encoding='utf8') as f:

    for line in f:
        line = line.strip().split()
        embeddings_vecs[line[0].lower()] = line[1:]

embeddings_dim = len(embeddings_vecs[line[0]])

print('Found {} word embeddings'.format(len(embeddings_vecs)))
print('Embeddings dimension: {}'.format(embeddings_dim))
```

We need to create a vocabulary index in which each word is associated with an integer. We will use the Keras [Tokenizer](https://www.tensorflow.org/api_docs/python/tf/keras/preprocessing/text/Tokenizer) class. Later, this class will be used to transform our sentences into integers.

```python
tokenizer = text.Tokenizer(filters='\t\n')

# Update internal vocabulary based on the GloVe vocabulary
tokenizer.fit_on_texts(embeddings_vecs.keys())
```

We need to prepare a corresponding embedding matrix that can be used in a Keras [Embedding](https://www.tensorflow.org/api_docs/python/tf/keras/layers/Embedding) layer. It's a simple NumPy matrix where the ith row is the pre-trained vector for the word of index i in our tokenizer.

```python
max_index = max(tokenizer.index_word.keys())

embedding_mat = np.zeros((max_index + 1, embeddings_dim))

for idx, word in tokenizer.index_word.items():
    embedding_mat[idx] = embeddings_vecs[word]
```

## Build and tune the model

We need to convert our sentences into sequences of integers and the class vectors to binary class matrixes. Also, most deep learning frameworks require that all sequences in the same batch have the same length. So we need to pad our sentences.

```python
X_train_s = tokenizer.texts_to_sequences(X_train)
X_test_s = tokenizer.texts_to_sequences(X_test)

max_len = max([len(val) for val in X_train_s + X_test_s])

print('Maximum sentence length: ', max_len)

X_train_s = sequence.pad_sequences(X_train_s, max_len, padding='post')
X_test_s = sequence.pad_sequences(X_test_s, max_len, padding='post')

Y_train_oh = keras.utils.to_categorical(Y_train, num_classes=5)
```

We will use the following architecture. The Long Short-Term Memory (LSTM) layers allow the model to account for word dependencies.

<img src='Model.png' style="width:700px">

```python
def create_model(units=64, dropout_rate=0, learning_rate=0.001):
    '''Creates a model using the specified hyperparameters.'''

    # Initialize the model
    model = keras.Sequential()

    # Add layers
    initializer = keras.initializers.HeNormal()

    model.add(Embedding(*embedding_mat.shape, embeddings_initializer=keras.initializers.Constant(embedding_mat),
                        input_length=max_len, trainable=False))
    model.add(LSTM(units=units, kernel_initializer=initializer, return_sequences=True))
    model.add(Dropout(dropout_rate))
    model.add(LSTM(units=units, kernel_initializer=initializer))
    model.add(Dropout(dropout_rate))
    model.add(Dense(5, activation='softmax'))

    # Compile the model
    model.compile(keras.optimizers.Adam(learning_rate), 'categorical_crossentropy', ['accuracy'])

    return model
```

We will start by tuning the initial value of the learning rate and keep that value constant during training.

```python
# Tune the initial learning rate
fit_param = {'X_train':X_train_s,'Y_train':Y_train_oh,'epochs':300}

compare_keras_param(create_model, {}, fit_keras_model, fit_param, ('learning_rate',[1E-4,5E-4,1E-3,5E-3,1E-2]))
```

We will use 0.0005 for the initial learning rate. Let's tune the learning rate's decay rate to get a smoother variation. Also, using 150 epochs to train the model seems enough.

```python
create_param = {'learning_rate':0.0005}
fit_param['epochs'] = 150

# Tune the decay rate of the learning rate
vals = [1,0.999,0.998]
schedulers = [[LearningRateScheduler(lambda epoch,lr,val=val: lr*val**epoch)] for val in vals]

compare_keras_param(create_model, create_param, fit_keras_model, fit_param, ('callbacks',schedulers),
                    [str(val) for val in vals])
```

Not using a decay rate leads to the best results. Now let‚Äôs tune the batch size.

```python
# Tune the batch size
compare_keras_param(create_model, create_param, fit_keras_model, fit_param, ('batch_size',[16,32,64]))
```

A batch size of 16 yields the best results. We will now tune the number of units of the LSTM layers and the dropout layers' rates.

```python
fit_param['batch_size'] = 16

# Tune the number of units
fit_param['validation_split'] = 0.3

compare_model_param(create_model, create_param, fit_keras_model, fit_param, ('units',[16,32,64,128]))
```

```python
# Tune the dropout layers' rates
create_param['units'] = 64
vals = np.arange(0,0.7,0.1)

compare_model_param(create_model, create_param, fit_keras_model, fit_param, ('dropout_rate',np.arange(0,0.7,0.1)))
```

We will use 64 units for each LSTM layer and a rate of 0.2 for the dropout layers.

```python
create_param['dropout_rate'] = 0.5
fit_param['validation_split'] = 0

model = create_model(**create_param)
train_accuracy = fit_keras_model(model, **fit_param)[0]
```

## Analyze predictions

Now let‚Äôs calculate the accuracy on the test set.

```python
predictions = model.predict(X_test_s).argmax(axis=1)

accuracy = accuracy_score(Y_test, predictions)
print('Test accuracy: {:.2%}'.format(accuracy))
```

An accuracy of 89.29% in the test set seems a good result since our training set is very small. Let‚Äôs look at the wrong predictions and the confusion matrix to evaluate the results in more detail. Since the test set does not have a uniform distribution of labels, we will normalise the confusion matrix over the rows.

```python
wrong_label = Y_test != predictions

for idx in wrong_label[wrong_label].index:
    print('{}; True:{}/Predicted:{}'.format(X_test[idx].strip(), emoji_dict[Y_test[idx]], emoji_dict[predictions[idx]]))  
```

```python
ConfusionMatrixDisplay(confusion_matrix(Y_test, predictions, normalize='true')).plot();
```

There is no significant difference in accuracy between the different labels, and the inaccurate results do not seem to be related to long-term dependencies. Let‚Äôs analyse the influence of the training set‚Äôs size on the model‚Äôs accuracy.

```python
model = create_model(**create_param)
model.save_weights('model.h5')

accuracies = []
sizes = range(33,133,33)

for size in sizes:

    fit_keras_model(model, X_train_s[:size], Y_train_oh[:size], epochs=150, batch_size=16, print_info=False)

    accuracies.append(accuracy_score(Y_test, model.predict(X_test_s).argmax(axis=1)))
    
    model.load_weights('model.h5')

os.remove('model.h5')

plt.plot(sizes, accuracies)
plt.xlabel('Size of training set')
plt.ylabel('Accuracy')
set_spines()
```

Having more data to train the model would most likely improve our model‚Äôs accuracy.
